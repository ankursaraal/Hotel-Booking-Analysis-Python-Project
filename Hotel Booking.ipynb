# Data analysis and visualisation 

```python
canceled_percentage = df['is_canceled'].value_counts(normalize=True)
print(canceled_percentage)
#37% reservation is canceled
```

```python
plt.figure(figsize=(4,3))
plt.bar(['not canceled', 'canceled'],df['is_canceled'].value_counts())
plt.title("Reservation Status Count")
plt.show
```

```python
# Plot the data
plt.figure(figsize=(4,3))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df)

# Set the legend
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(legend_labels, ['not canceled', 'canceled'], bbox_to_anchor=(1,1))

# Set the title and labels
plt.title("Reservation status in different Hotels", size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')

# Show the plot
plt.show()
```

```python
resort_hotel = df[df['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
#27% cancellation of bookings
```

```python
city_hotel = df[df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
#41% cancellation of bookings
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average daily rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize =(16,8))
ax1 = sns.countplot(x='month', hue = 'is_canceled', data = df)
legend_label,_= ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per Month', size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python

df['is_canceled'] = df['is_canceled'].astype(int)

# Create a new DataFrame with the grouped and summed data
filtered_data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()

plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
sns.barplot(x='month', y='adr',data=filtered_data,)
plt.show()
```

```python
canceled_data = df[df['is_canceled']==1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```
